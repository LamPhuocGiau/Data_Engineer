```python
# import matplotlib and pandas
from matplotlib import pyplot as plt
import pandas as pd

# load dataset
scatter_data = pd.read_csv('spotify_data_by_genres.csv')
```

```python
# display charts correctly
%matplotlib inline
plt.rcParams['figure.figsize'] = (5, 3)
plt.rcParams['figure.dpi'] = 75
```

2. Write the code to make a scatterplot comparing **danceability** and **valence**. Danceability is a measure of how “danceable” a song is, from 0 (not dance-y, like an audiobook or funeral dirge) to 1 (very dance-y, like disco or house music). Valence is a measure of the song’s mood: sad to happy from 0 to 1. 

```python
## YOUR SOLUTION HERE ##
plt.scatter(scatter_data.danceability, scatter_data.valence)

plt.title('Mood and Danceability correlation in Spotify genres')
plt.xlabel('Danceability')
plt.ylabel('Valence / Mood (sadder to happier)')
plt.show()
```

3. Adjust the `alpha` to 0.25 to make the relationship more obvious. Play around with other alpha values to see how the graph’s readability changes! 

```python
## YOUR SOLUTION HERE ##
plt.scatter(scatter_data.danceability, scatter_data.valence, alpha = 0.25)

plt.title('Mood and Danceability correlation in Spotify genres')
plt.xlabel('Danceability')
plt.ylabel('Valence / Mood (sadder to happier)')
plt.show()
```

4. Change the `color` to ‘teal’ and the `alpha` to 0.15.

```python
## YOUR SOLUTION HERE ##
plt.scatter(scatter_data.danceability, scatter_data.valence, color = 'teal', alpha = 0.15)

plt.title('Mood and Danceability correlation in Spotify genres')
plt.xlabel('Danceability')
plt.ylabel('Valence / Mood (sadder to happier)')
plt.show()
```
